```python
# Define the function and its derivative
def func(x):
    return (x + 3)**2

def derivative(x):
    return 2 * (x + 3)

# Implement Gradient Descent
def gradient_descent(starting_x, learning_rate, num_iterations):
    x = starting_x  # Starting point
    x_values.append(x)
    y_values.append(func(x))

    for i in range(num_iterations):
        grad = derivative(x)              # Compute gradient
        x = x - learning_rate * grad      # Update x
        x_values.append(x)
        y_values.append(func(x))
        print(f"Iteration {i+1}: x = {x}, f(x) = {func(x)}")

    return x

# Parameters
starting_x = 2
learning_rate = 0.1
num_iterations = 50

# Run Gradient Descent
min_x = gradient_descent(starting_x, learning_rate, num_iterations)
print(f"\nLocal minimum occurs at x = {min_x}")
```

```python
import matplotlib.pyplot as plt
# Visualization
plt.plot(x_values, y_values, marker='o')
plt.title("Convergence of Gradient Descent")
plt.xlabel("x values")
plt.ylabel("f(x) values")
plt.grid(True)
plt.show()
```
